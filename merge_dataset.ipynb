```python
import os

folder_path = 'dataVisualizationCaseStudy\city_listings'

list_of_files = []
for item in os.listdir(folder_path):
    item_path = os.path.join(folder_path, item)
    if os.path.isfile(item_path):
        list_of_files.append(item)

print(list_of_files)




```

```python
import os
import pandas as pd

folder_path = 'dataVisualizationCaseStudy\city_listings'

dataframe = []
for item in os.listdir(folder_path):
    item_path = os.path.join(folder_path, item)
    if os.path.isfile(item_path):
        df = pd.read_csv(item_path)
        dataframe.append(df)

complete = pd.concat(dataframe, ignore_index=True)

print(dataframe)
print(complete)

```
