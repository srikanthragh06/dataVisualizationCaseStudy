```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import os
```

- In city_listings folder, takes all the csv files turns them into dataframes appends them into a dataframe list 
- We row append all the dataframes in the dataframe list
- We upload the dataframe in a csv file

```python
folder_path = 'city_listings'

dataframes = []
for item in os.listdir(folder_path):
    item_path = os.path.join(folder_path, item)
    if os.path.isfile(item_path):
        df = pd.read_csv(item_path)
        dataframes.append(df)

complete = pd.concat(dataframes, ignore_index=True)

complete.to_csv('./American_house_prices_dataset.csv')

complete
```

- We find the missing value % in all columns of a dataframe.
- We delete the columns that have more than 20% missing values.
- Columns with a large proportion of missing values may indicate poor data quality or inconsistency.

```python
missingPerc=df.isnull().sum()/df.shape[0]
missingPerc
```

```python
toomuchmissingCols=missingPerc[missingPerc>0.2].index
toomuchmissingCols
```

```python
df.drop(columns=toomuchmissingCols,inplace=True)
```

```python
df
```

- There are 4 columns which all indicate price
- we will rename hdpData.homeInfo.price and drop the rest of the columns

```python
priceColumns=[]
for col in df.columns:
    if "price" in col:
        priceColumns.append(col)
priceColumns
```

```python
df[priceColumns]
```

```python
df.drop(columns=["price","priceLabel","hdpData.homeInfo.priceForHDP"],inplace=True)
df.rename(columns={"hdpData.homeInfo.price":"price"},inplace=True)
```

```python
df
```

Since price is our target feature, we have to delete rows that have price as null

```python
nullPriceIndices=df["price"][df["price"].isnull()==True].index
df.drop(nullPriceIndices,inplace=True)
df.reset_index(inplace=True)
```

```python
df
```

Features that have a category which is more than 97.5% proportion have been deleted as they are not useful

```python
tooMuch1Category=[]
for col in df.columns:
    maxVal=df[col].value_counts().sort_values(ascending=False).index[0]

    maxValProportion=df[col][df[col]==maxVal].shape[0]/df[col].shape[0]
    if maxValProportion>0.975:
        print(col,maxVal,maxValProportion*100)
        tooMuch1Category.append(col)

```

```python
print(f"We have {len(tooMuch1Category)} such columns")
```

```python
df.drop(columns=tooMuch1Category,inplace=True)
df
```

We reviewed each column thoroughly and removed those that we felt did not contribute significantly to predicting the price.

Deleting index, Unnamed: 0.1, Unnamed: 0, zpid because they are unnecesary features

```python
deletingColsDf=df[['index', 'Unnamed: 0.1', 'Unnamed: 0', 'zpid', 'imgSrc']]
df.drop(columns=['index', 'Unnamed: 0.1', 'Unnamed: 0', 'zpid', 'imgSrc'], inplace=True)
deletingColsDf
```

Deleting 'hdpData.homeInfo.latitude', 'hdpData.homeInfo.longitude','hdpData.homeInfo.bathrooms','hdpData.homeInfo.bedrooms', 'hdpData.homeInfo.livingArea','hdpData.homeInfo.zpid' as they are  duplicate features

Deleting hdpData.homeInfo.dateSold as it is not a date and we couldn't find any interpretations from it

```python
deletingColsDf = df[['hdpData.homeInfo.dateSold']]

df.drop(columns=['hdpData.homeInfo.dateSold'], inplace=True)

deletingColsDf
```

Deleting address,detailUrl as the information in it is already present in other features like streetName and city name

```python
deletingColsDf = df[['address', "detailUrl"]]

df.drop(columns=["address", "detailUrl"], inplace=True)

deletingColsDf
```

Deleting street address, zipcode as each row is unique

```python
deletingColsDf = df[['hdpData.homeInfo.streetAddress', 'hdpData.homeInfo.zipcode']]

df.drop(columns=['hdpData.homeInfo.streetAddress', 'hdpData.homeInfo.zipcode'], inplace=True)

deletingColsDf
```

Deleting variableData.text as all the data here is within 2 months and the date is not useful

```python
def removeColPrefix(colName):
    prefixes = ["hdpData.homeInfo.", "latLong."]
    for prefix in prefixes:
        if colName.startswith(prefix):
            colName = colName.replace(prefix, "")
            break  # Exit loop once a prefix is found and removed
    return colName


df.rename(columns={key:removeColPrefix(key) for key in df.columns },inplace=True)

df
```
