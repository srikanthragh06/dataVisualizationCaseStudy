```python
def barforcatdata(data,df):
  plt.figure(figsize=(20,20))
  barplots=[]
  for i in df.columns:
    countpercent=(df[data].value_counts()/df[i].shape[0])*100
    barplot=plt.bar(x=countpercent.index,height=countpercent.values)
    barplots.append(barplot)
  return barplots


for j in catogricalcol:
  barplots=barforcatdata(j,df)
  for k in barplots:
    plt.show()
```
